### 练习：定义衡量标准
如果不能对模型的训练和测试的表现进行量化地评估，我们就很难衡量模型的好坏。通常我们会定义一些衡量标准，这些标准可以通过对某些误差或者拟合程度的计算来得到。在这个项目中，你将通过运算[*决定系数*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination) R<sup>2</sup> 来量化模型的表现。模型的决定系数是回归分析中十分常用的统计信息，经常被当作衡量模型预测能力好坏的标准。

R<sup>2</sup>的数值范围从0至1，表示**目标变量**的预测值和实际值之间的相关程度平方的百分比。一个模型的R<sup>2</sup> 值为0还不如直接用**平均值**来预测效果好；而一个R<sup>2</sup> 值为1的模型则可以对目标变量进行完美的预测。从0至1之间的数值，则表示该模型中目标变量中有百分之多少能够用**特征**来解释。_模型也可能出现负值的R<sup>2</sup>，这种情况下模型所做预测有时会比直接计算目标变量的平均值差很多。_

在下方代码的 `performance_metric` 函数中，你要实现：
- 使用 `sklearn.metrics` 中的 `r2_score` 来计算 `y_true` 和 `y_predict`的R<sup>2</sup>值，作为对其表现的评判。
- 将他们的表现评分储存到`score`变量中。

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score
def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

### 问题 7- 网格搜索（Grid Search）
*什么是网格搜索法？如何用它来优化学习算法？*


**回答: 
网格搜索法是把预先给定的数据训练方法和参数进行组合，并逐一代入算法运行评估得分，最后通过比较得分高低获得最佳的训练方法和参数组合的方法。由于它返回了最佳的训练方法和参数组合，我们可以藉由它得到能输出更好结论的模型，使得模型能以较高相关性地代表普适性数据，且偏差和方差都得以控制在折中最优的状态。

回复“网格搜索只是搜索单一算法下表现最好的参数组合。它本身并不包含对算法的筛选。另外你还提到了“它通过过比较得分高低来获得最佳参数组合”，能详细说一下分数是怎样得出的吗？”：
我认为整个网格搜索算法而言，其实准确来说应该算是一个搜索策略吧，搜索这个词语本身就包含两个层面（第一层是算法实现了尝试不同组合的功能，第二层是所采用的不同评分目标的机制本身，这里评分机制就是我所表达的训练方法(通过设定具体的评价目标而导致了不同的训练过程),机制本身也算是一种算法，因为它是实现了具体功能的，只不过在这里它恰好作为参数代入到更宏观影响层面的算法当中而已）。
回到宏观层面而言，其实整个网格搜索算法应该做了几个事情：(1）尝试不同参数组合；（2）根据实际参数当中的评分机制参数（训练方法），协同其他参数算出具体的分值；【算出分数其实就是评分机制本身，以特定的权重奖罚各个输入数据当中呈现的表征，进而通过表征之间的数学比例关系转化为以比例为依据的纯粹数字运算问题，就可以得出分数了】（3）比较各个参数组合的具体分值找出最大的那个。（实际上就是筛选过程）

- 优化模型时，使用这种方法对网格搜索有什么好处？网格搜索时如果不使用交叉验证会有什么问题？为什么？

**提示：** 在下面 fit_model函数最后加入 `print pd.DataFrame(grid.cv_results_)` 可以帮你查看更多。

**答案: 
K-fold cross-validation是一种优化的训练集、测试集分离应用方法。它把样本先分为训练集和测试集。（从参数设置可见）然后训练集再进一步K等分，使得从第二份样本开始，每份样本作为训练集的同时也将作为对前面训练集的验证集【等分后的验证集要与前面提取出来的测试集作概念区别】，验证集通过计算当前验证分数，并与过去验证分数一起取平均分（若有），持续总共得到K次分数，持续反馈，让算法重新判断各个参数组合的最高得分，从而不断采取最优参数（如何修改参数见前面问题7的回答），进而使得学习算法的评估更为准确。

网格搜索生成一个网格，对算法参数进行不同组合尝试，交叉验证则返回模型评估值为网格搜索的不同组合尝试提供评价（评分）依据，最后，根据网格内不同的尝试组合各自的直观评价依据，进行比照，可获得最佳参数组合。【评分依据的生成，见问题7的回答】

使用K-fold cross-validation作为评估方法，可使得网格搜索找到使得模型复杂度较为恰当、使得模型随着处理数据量越大，显得更有效的模型参数。

训练与测试目的是使得模型尽可能地描述同一个类型的全体统一性数据，而要得到这样描述能力的参数需要足够大量的数据支持。在达成这个目的的过程中，每个训练节点得到的参数都会面临欠拟合或过拟合的问题，数据越多，这种偏差调整的范围基于前面所有的测试，会逐步地缩小，偏差数值逐渐收敛，方差自然会随之逐步缩小。

单纯的使用训练集和测试集，只提供了1个测试节点的结果用于调整拟合水平，而使用交叉验证，则提供了多个测试节点的结果用于调整拟合水平。
在数据量不大的时候，单独1次调整拟合与多次调整拟合的差异还不能彻底的呈现出来。数据量越大，它们的偏差、方差差异会逐渐呈现出来。

'cv_results'返回给我们的是分类器测试的最优参数
'best_estimator'返回的是携带着最优参数的估计器空壳。

### 练习：训练模型
在最后一个练习中，你将需要将所学到的内容整合，使用**决策树演算法**训练一个模型。为了保证你得出的是一个最优模型，你需要使用网格搜索法训练模型，以找到最佳的 `'max_depth'` 参数。你可以把`'max_depth'` 参数理解为决策树算法在做出预测前，允许其对数据提出问题的数量。决策树是**监督学习算法**中的一种。

此外，你会发现你的实现使用的是 `ShuffleSplit()` 。它也是交叉验证的一种方式（见变量 `'cv_sets'`）。虽然这不是**问题8**中描述的 K-Fold 交叉验证，这个教程验证方法也很有用！这里 `ShuffleSplit()` 会创造10个(`'n_splits'`)混洗过的集合，每个集合中20%(`'test_size'`)的数据会被用作**验证集**。当你在实现的时候，想一想这跟 K-Fold 交叉验证有哪些相同点，哪些不同点？

在下方 `fit_model` 函数中，你需要做的是：
- 使用  `sklearn.tree` 中的 [`DecisionTreeRegressor`](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html) 创建一个决策树的回归函数；
  - 将这个回归函数储存到 `'regressor'` 变量中；
- 为 `'max_depth'` 创造一个字典，它的值是从1至10的数组，并储存到 `'params'` 变量中；
- 使用 `sklearn.metrics` 中的 [`make_scorer`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html)  创建一个评分函数；
  - 将 `performance_metric` 作为参数传至这个函数中；
  - 将评分函数储存到 `'scoring_fnc'` 变量中；
- 使用 `sklearn.model_selection` 中的 [`GridSearchCV`](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) 创建一个网格搜索对象；
  - 将变量`'regressor'`, `'params'`, `'scoring_fnc'`, 和 `'cv_sets'` 作为参数传至这个对象中；
  - 将 `GridSearchCV` 存到 `'grid'` 变量中。
  
如果有同学对python函数如何传递多个参数不熟悉，可以参考这个MIT课程的[视频](http://cn-static.udacity.com/mlnd/videos/MIT600XXT114-V004200_DTH.mp4)。

```python
# TODO: Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.metrics import make_scorer
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    cv_sets = ShuffleSplit(n_splits=10, test_size=0.20, random_state = 0,train_size=None)

    # TODO: Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth':range(1,11)}

    # TODO: Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # TODO: Create the grid search object
    grid = GridSearchCV(regressor,params,scoring_fnc,cv=cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    #print pd.DataFrame(grid.cv_results_)
    return grid.best_estimator_
fit_model(features, prices)
```

- School: 是否为学区房，0或1

```python
#Part 1
# 载入此项目所需要的库
import numpy as np
import pandas as pd
import visuals as vs # Supplementary code

# Pretty display for notebooks
# 让结果在notebook中显示
%matplotlib inline

# 载入北京房屋的数据集
data = pd.read_csv('bj_housing.csv')
#print data
prices = data['Value']
features = data.drop('Value', axis = 1)
#print features
# 完成
print "Beijing housing dataset has {} data points with {} variables each.".format(*data.shape)

#***part 2 数据分割与重排要加上（测试集不要和后面的训练集中的验证集概念相混淆）***
# Import 'train_test_split'
from sklearn.model_selection import train_test_split

# Shuffle and split the data into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.20, random_state=42)

##由于这里特征数变大，params不应该参数不应该只设置到1-10，可以设置多一点，要看图
#vs.ModelLearning(features, prices)
#vs.ModelComplexity(X_train, y_train)

#part 3
from sklearn.metrics import r2_score
def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score

#part 4
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.model_selection import ShuffleSplit

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    cv_sets = ShuffleSplit(n_splits=10, test_size=0.2, random_state = 42,train_size=None)

    # Create a decision tree regressor object若不确定一个random_state,预测度会浮动。要看图设置max_depth
    regressor = DecisionTreeRegressor(max_depth=3)

    # Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth':range(1,11)}

    # Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # Create the grid search object
    grid = GridSearchCV(regressor,params,scoring_fnc,cv=cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    #print pd.DataFrame(grid.cv_results_)
    return grid.best_estimator_

#part 5
# Fit the training data to the model using grid search
reg = fit_model(X_train, y_train)

# Produce the value for 'max_depth'

print "Parameter 'max_depth' is {} for the optimal model.".format(reg.get_params()['max_depth'])
print performance_metric(y_test, reg.predict(X_test))
#print reg.predict(X_test)
```
